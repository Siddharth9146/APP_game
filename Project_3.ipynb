```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.yes_link = None
        self.no_link = None

def loadGame(file_name):
    root = None
    try:
        with open(file_name, "r") as file:
            lines = file.readlines()
            game_title = lines[0].strip()
            help_info = lines[1].strip()
            data_lines = lines[2:]

        for line in data_lines:
            node_number, question_or_answer = line.strip().split(" ", 1)
            node = TreeNode(question_or_answer)
            if root is None:
                root = node
            else:
                buildTree(root, node, int(node_number))
    except FileNotFoundError:
        return None

    return root, game_title, help_info

def buildTree(current_node, node_to_insert, node_number):
    if node_number % 10 == 0:
        if current_node.no_link is None:
            current_node.no_link = node_to_insert
        else:
            buildTree(current_node.no_link, node_to_insert, node_number)
    else:
        if current_node.yes_link is None:
            current_node.yes_link = node_to_insert
        else:
            buildTree(current_node.yes_link, node_to_insert, node_number)

def playGame(current_node):
    while current_node:
        print(current_node.data)
        answer = input("Is it (Y/N)? ").strip().lower()
        while answer not in ["y", "n", "yes", "no"]:
            print("Invalid input. Please enter 'Y' or 'N'.")
            answer = input("Is it (Y/N)? ").strip().lower()

        if answer in ["y", "yes"]:
            current_node = current_node.yes_link
        else:
            current_node = current_node.no_link

def displayTreeInorder(node):
    if node:
        displayTreeInorder(node.yes_link)
        print(node.data)
        displayTreeInorder(node.no_link)

if __name__ == "__main__":
    game_file = "game1.txt"
    game_data = loadGame(game_file)

    if game_data:
        root, game_title, help_info = game_data
        while True:
            print("\n" + game_title)
            print("P: Play the game")
            print("L: Load another game file")
            print("D: Display the binary tree")
            print("H: Help information")
            print("X: Exit the program")
            choice = input("...your choice: ").strip().lower()

            if choice == "p":
                playGame(root)
            elif choice == "l":
                new_game_file = input("Enter the game file name: ")
                new_game_data = loadGame(new_game_file)
                if new_game_data:
                    root, game_title, help_info = new_game_data
                else:
                    print("Failed to load the new game file.")
            elif choice == "d":
                print("What order do you want to display?")
                print("1. Inorder")
                print("2. Preorder")
                print("3. Postorder")
                print("4. Return to main menu")
                display_choice = input("...your choice: ").strip().lower()
                if display_choice == "1":
                    displayTreeInorder(root)
            elif choice == "h":
                print(help_info)
            elif choice == "x":
                print("Exiting the program.")
                break
            else:
                print("Invalid choice. Please choose a valid option.")
    else:
        print("Failed to load the game file.")
```
